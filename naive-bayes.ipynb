```python
from __future__ import division #To avoid integer division
from operator import itemgetter
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data_set.csv", low_memory=False)
```

```python
data = data[['reviews.rating', 'reviews.text']]
train, test = train_test_split(data, test_size=0.2)
```

```python
pos_rev = []
neg_rev = []
neut_rev = []
```

```python
# for i in range(len(data)):
#     if data[]
train['reviews.rating'] = pd.to_numeric(train['reviews.rating'], errors='coerce')
train = train.dropna(subset=['reviews.rating'])
train['reviews.rating'] = train['reviews.rating'].astype(int)

# a = 0
for col, row in train.iterrows():
    if int(row["reviews.rating"]) > 3:
        pos_rev.append(row["reviews.text"])
    elif int(row["reviews.rating"]) < 3:
        neg_rev.append(row["reviews.text"])
    else:
        neut_rev.append(row["reviews.text"])
# a
# neut_rev
```

```python
words = {}
pos_len = 0
neg_len = 0
neut_len = 0

# print (len(pos_rev))

for rev in pos_rev:
    try:
        spl_rev = rev.split()
        pos_len += 1
        for item in spl_rev:
            if item in words:
                words[item][0] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][0] += 1
    except:
        print("h")

# pos_len

for rev in neg_rev:
    try:
        spl_rev = rev.split()
        neg_len += 1
        for item in spl_rev:
            if item in words:
                words[item][1] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][1] += 1
    except:
        print("h")
        
for rev in neut_rev:
    try:
        spl_rev = rev.split()
        neut_len += 1
        for item in spl_rev:
            if item in words:
                words[item][2] += 1
            else:
                lis = [0] * 3
                words[item] = lis
                words[item][2] += 1
    except:
        print("h")
```

```python
test['reviews.rating'] = pd.to_numeric(test['reviews.rating'], errors='coerce')
test = test.dropna(subset=['reviews.rating'])
test['reviews.rating'] = test['reviews.rating'].astype(int)

denominator = len(test)
numerator = 0

def calc_pos(rev):
    to_return = 1
    flag = 0
    try:
        spl_rev = rev.split()
        for item in spl_rev:
            if item in words:
                flag = 1
                prob = (float) (words[item][0]) / (words[item][0] + words[item][1] + words[item][2])
                to_return *= prob
    except:
        print ("h")
    if flag == 1:
        return to_return
    return 0  # if all words are new and not in dictionary

def calc_neg(rev):
    to_return = 1
    flag = 0
    try:
        spl_rev = rev.split()
        for item in spl_rev:
            if item in words:
                flag = 1
                prob = (float) (words[item][1]) / (words[item][0] + words[item][1] + words[item][2])
                to_return *= prob
    except:
        print ("h")
    if flag == 1:
        return to_return
    return 0  # if all words are new and not in dictionary

def calc_neut(rev):
    to_return = 1
    flag = 0
    try:
        spl_rev = rev.split()
        for item in spl_rev:
            if item in words:
                flag = 1
                prob = (float) (words[item][2]) / (words[item][0] + words[item][1] + words[item][2])
                to_return *= prob
    except:
        print ("h")
    if flag == 1:
        return to_return
    return 0  # if all words are new and not in dictionary

pos_percent = 0
neg_percent = 0
neut_percent = 0

for col, row in test.iterrows():
    rev = row["reviews.text"]
    pos_val = calc_pos(rev)
    neg_val = calc_neg(rev)
    neut_val = calc_neut(rev)
    
    if pos_val >= neg_val and pos_val >= neut_val and int(row["reviews.rating"]) > 3:
        numerator += 1
        pos_percent += 1
    elif neg_val >= pos_val and neg_val >= neut_val and int(row["reviews.rating"]) < 3:
        numerator += 1
        neg_percent += 1
    elif neut_val >= neg_val and neut_val >= pos_val and int(row["reviews.rating"]) == 3:
        numerator += 1
        neut_percent += 1
        
accuracy = (float) (numerator) / (denominator)
# pos_percent = (float) (pos_percent) / (denominator)
# neg_percent = (float) (neg_percent) / (denominator)
# neut_percent = (float) (neut_percent) / (denominator)

print ("accuracy = ", accuracy * 100)
# print ("positive % = ", pos_percent * 100)
# print ("negative % = ", neg_percent * 100)
# print ("neutral % = ", neut_percent * 100)

    
```
